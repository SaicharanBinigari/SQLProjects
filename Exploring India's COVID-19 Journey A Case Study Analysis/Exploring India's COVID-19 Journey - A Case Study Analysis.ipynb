#### Data Cleaning for death_and_recovery dataset

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

# Task1
# Loading the data
def read_data_from_csv():
    df = pd.read_csv('death_and_recovery.csv')
    return df


def data_cleaning():
    # DO NOT REMOVE FOLLOWING LINE
    data = read_data_from_csv()

    # Remove Unwanted columns (comorbidity, State_code)
    unwanted_columns = ['comorbidity', 'State_code']
    data.drop(columns=unwanted_columns, inplace=True)

    # Only keep these columns in the dataset (Age, Gender, Patient_status, City, State)
    allowed_columns = ['Age', 'Gender', 'Patient_status', 'City', 'State']
    data = data[allowed_columns]

    # Export cleaned Dataset to a new CSV file named "death_and_recovery_cleaned.csv"
    data.to_csv('death_and_recovery_cleaned.csv', index=False)
    
    return data
```

```python
data_cleaning()
```

#### Data Cleaning for hospitalbeds dataset

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

#Task1
#Loading the data
def read_data_from_csv():
    #read the dataset
    df = pd.read_csv('hospitalbeds.csv')
    return df


#Task 2: Renaming the Columns
def rename_columns():
    data=read_data_from_csv()
    """Rename the columns in the dataset. ("'serial': 'sno','state': 'State_UT','Hosp_Aval': 'Hospitals_Available',
    'Beds_Aval': 'Beds_Available','Pop_beds':'Population_beds'")"""
    data.rename(columns={'serial': 'sno','state': 'State_UT','Hosp_Aval': 'Hospitals_Available',
    'Beds_Aval': 'Beds_Available','Pop_beds':'Population_beds'}, inplace=True)


    # export cleaned Dataset to newcsv file named "hospitalbeds_cleaned.csv"
    data.to_csv("hospitalbeds_cleaned.csv", index = False)
    return data
```

```python
rename_columns()
```

#### Data Cleaning for statewisedata dataset

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

#Task1
#Loading the data
def read_data_from_csv():
    df = pd.read_csv('statewisedata.csv')
    return df


def data_cleaning():
    # DO NOT REMOVE FOLLOWING LINE
    data = read_data_from_csv()

    # Remove Unwanted columns(Delta_Confirmed,Delta_Recovered,Delta_Deaths)
    unwanted_columns = ['Delta_Confirmed','Delta_Recovered','Delta_Deaths']
    data.drop(columns=unwanted_columns, inplace=True)
    # only these columns are allowed in the dataset
    #(State_UT,Confirmed,Recovered,Deaths,Active,Last_updad_time,State_code)
    allowed_columns = ['State_UT','Confirmed','Recovered','Deaths','Active','Last_updad_time','State_code']
    data = data[allowed_columns]
    # export cleaned Dataset to newcsv file named "statewisedata_cleaned.csv"
    data.to_csv("statewisedata_cleaned.csv", index = False)
    return data
```

#### Data Cleaning for agedistribution_2016_estimates dataset

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

#Task1
#Loading the data
def read_data_from_csv():
    #Read the dataset
    df = pd.read_csv('agedistribution_2016_estimates.csv')
    return df


#Task 2: Renaming the Columns
def rename_columns():
    data=read_data_from_csv()
    """(Rename the column names as 'Age': 'Age_group',
        'M': 'Male',
        'F': 'Female',
        'Tot': 'Total')"""
    data.rename(columns={'Age': 'Age_group',
        'M': 'Male',
        'F': 'Female',
        'Tot': 'Total'}, inplace=True)
    

    # export cleaned Dataset to newcsv file named "agedistribution_2016_estimates_cleaned.csv"
    data.to_csv("agedistribution_2016_estimates_cleaned.csv", index = False)
    
    return data
```
